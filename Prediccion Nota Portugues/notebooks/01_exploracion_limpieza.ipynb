Realizamos analisis de correlaciones y algunas distribuciones ya que es necesario para realiar nuestra regresion lineal para poder predecir la nota G3

```python
#Librerías necesarias para la graficación de nuestros datos
import matplotlib.pyplot as plt 
import seaborn as sns

numericas = df.select_dtypes(include=['int64', 'float64']).columns.to_list() #Seleccionamos solo las variables numericas
sns.pairplot(df[numericas]) #Cargamos un pairplot que nos va dar correlaciones entre las variables que tiene numerica y sus distribuciones
```

Con el grafico podemos evidenciar que G1 G2 son las que mas se correlacionan con la variable  G3 y Absences un poco lo cual da buenos indicios para utilizar solamente esas variables como inicio ya que el modelo de regresion lineal funciona mucho mejor a pocas variables 

Vamos a realizar una grafica  con todas las variables en general para poder revisar incluso las variables categoricas si tenemos o no correlación con la variable objetivo y rectificiar las 3 variables pasadas si están correlacionadas entre si 

```python
#Creamos una copia del dataframe ya que relizaremos un dummies el cual modifica las variables originales entonces para dejar el dataframe original sin modificar 
df_copy = df.copy()

df_copy = pd.get_dummies(df_copy, drop_first=True, dtype=int) #Eliminamos  algunas variables repetidas para evitar multicolonealidad ya que las regresión se ven algo afectadas por eso
```

```python
#Realizamos la graficacion
plt.figure(figsize=(14, 8))
df_copy.corr()['G3'].sort_values(ascending=False).plot(kind='bar')
plt.xticks(rotation=90)
plt.show()
```

```python
df_copy.corr()['G3'] #Revisamos numericamente cada valor de correlacion respecto a G3(Variable objetivo)
```
